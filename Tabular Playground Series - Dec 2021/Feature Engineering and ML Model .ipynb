## As we know about that Aspect is in degrees azimuth, we can try to shift it at 180

```python
def r(x):
    if x+180>360:
        return x-180
    else:
        return x+180
    
df['Aspect2'] = df.Aspect.map(r)
df_test['Aspect2'] = df_test.Aspect.map(r)
```

```python
# out of first 10 columns most of them have some negative values
```

## Finally, we can create some features from those which have similar meanings, f.e. distances

```python
df['Distanse_to_Hydrolody'] = (df['Horizontal_Distance_To_Hydrology']**2+df['Vertical_Distance_To_Hydrology']**2)**0.5
df_test['Distanse_to_Hydrolody'] = (df_test['Horizontal_Distance_To_Hydrology']**2+df_test['Vertical_Distance_To_Hydrology']**2)**0.5

df['Hydro_Fire_1'] = abs(df['Horizontal_Distance_To_Hydrology'])+abs(df['Horizontal_Distance_To_Fire_Points'])
df_test['Hydro_Fire_1'] = abs(df_test['Horizontal_Distance_To_Hydrology'])+abs(df_test['Horizontal_Distance_To_Fire_Points'])

df['Hydro_Fire_2'] = abs(df['Horizontal_Distance_To_Hydrology'])+abs(df['Horizontal_Distance_To_Fire_Points'])
df_test['Hydro_Fire_2'] = abs(df_test['Horizontal_Distance_To_Hydrology'])+abs(df_test['Horizontal_Distance_To_Fire_Points'])

df['Hydro_Road_1'] = abs(df['Horizontal_Distance_To_Hydrology'])+abs(df['Horizontal_Distance_To_Roadways'])
df_test['Hydro_Road_1'] = abs(df_test['Horizontal_Distance_To_Hydrology'])+abs(df_test['Horizontal_Distance_To_Roadways'])

df['Hydro_Road_2'] = abs(df['Horizontal_Distance_To_Hydrology'])+abs(df['Horizontal_Distance_To_Roadways'])
df_test['Hydro_Road_2'] = abs(df_test['Horizontal_Distance_To_Hydrology'])+abs(df_test['Horizontal_Distance_To_Roadways'])

df['Fire_Road_1'] = abs(df['Horizontal_Distance_To_Fire_Points'])+abs(df['Horizontal_Distance_To_Roadways'])
df_test['Fire_Road_1'] = abs(df_test['Horizontal_Distance_To_Fire_Points'])+abs(df_test['Horizontal_Distance_To_Roadways'])

df['Fire_Road_2'] = abs(df['Horizontal_Distance_To_Fire_Points'])+abs(df['Horizontal_Distance_To_Roadways'])
df_test['Fire_Road_2'] = abs(df_test['Horizontal_Distance_To_Fire_Points'])+abs(df_test['Horizontal_Distance_To_Roadways'])
```

```python
from sklearn import ensemble
```

```python
feature_cols = [col for col in df.columns if col not in ['Cover_Type','Id']]

X_train = df[feature_cols]
X_test = df_test[feature_cols]
y = df['Cover_Type']
test_ids = df_test['Id']

forest = ensemble.ExtraTreesClassifier(n_estimators=100, criterion='gini', max_depth=None, 
    min_samples_split=2, min_samples_leaf=1, max_features='auto', 
    bootstrap=False, oob_score=False, n_jobs=-1, random_state=None, verbose=0)

```

```python
forest.fit(X_train, y)

with open('features_engineering_benchmark.csv', "wb") as outfile:
    outfile.write("Id,Cover_Type\n")
    for e, val in enumerate(list(forest.predict(X_test))):
        outfile.write("%s,%s\n"%(test_ids[e],val))

```
